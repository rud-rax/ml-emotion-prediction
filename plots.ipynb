```python
df = pd.read_csv('re.csv')
data = df[['tonal_hcdf_std']]
```

```python
df[df['tonal_hcdf_std'] < 0]
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew ,  shapiro, ks_2samp , boxcox
from sklearn.preprocessing import PowerTransformer


# data_transformed, lambda_bc = boxcox(data + 1e-6)  # Add small constant to avoid zero


# Apply Yeo-Johnson transformation
pt = PowerTransformer(method='yeo-johnson')
data_transformed = pt.fit_transform(data)



# Plot distributions before and after transformation
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(data, bins=30, kde=True, ax=axes[0])
axes[0].set_title("Original Column Distribution")

sns.histplot(data_transformed, bins=30, kde=True, ax=axes[1])
axes[1].set_title("Yeo-Johnson Transformed Distribution")

plt.show()

# Skewness before and after transformation
print("Before Normalization: Skewness =", skew(data), "Kurtosis =", kurtosis(data))
print("After Normalization: Skewness =", skew(data_transformed), "Kurtosis =", kurtosis(data_transformed))


shapiro_test = shapiro(data_transformed)
print("Shapiro-Wilk Test: p-value =", round(shapiro_test.pvalue , 3))

# Compare original vs normalized (if normality is expected)
ks_test = ks_2samp(data, data_transformed)
print("KS Test: p-value =", ks_test.pvalue)

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew ,  shapiro, ks_2samp , boxcox



data_transformed, lambda_bc = boxcox(data.squeeze() + 1e-6)  # Add small constant to avoid zero


# Plot distributions before and after transformation
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(data, bins=30, kde=True, ax=axes[0])
axes[0].set_title("Original Column Distribution")

sns.histplot(data_transformed, bins=30, kde=True, ax=axes[1])
axes[1].set_title("Box Cox Transformed Distribution")

plt.show()

# Skewness before and after transformation
print("Before Normalization: Skewness =", skew(data), "Kurtosis =", kurtosis(data))
print("After Normalization: Skewness =", skew(data_transformed), "Kurtosis =", kurtosis(data_transformed))


shapiro_test = shapiro(data_transformed)
print("Shapiro-Wilk Test: p-value =", round(shapiro_test.pvalue , 3))

# Compare original vs normalized (if normality is expected)
ks_test = ks_2samp(data.squeeze(), data_transformed)
print("KS Test: p-value =", ks_test.pvalue)

```
